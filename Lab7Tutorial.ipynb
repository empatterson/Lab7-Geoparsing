```python
#This script will scrape Innocents Abroad from Project Gutenburg website
url = 'http://www.gutenberg.org/files/3176/3176-0.txt'
response = request.urlopen(url)
raw = response.read().decode('utf8')
#This will print a sample of the scraped book
print(f'{type(raw)}, \n{len(raw)}, \n{raw[:501]}')
```

```python
#This section uses GeoText to pull out and print city names from the text
places = GeoText (raw)
cities = list(places.cities)
cities
```

```python
#This script takes the city names and adds lat, long to them
geolocator = Nominatim(user_agent='empattersonclarku@gmail.com', timeout=2)

lat_lon = []
for city in cities:
  try:
    location = geolocator.geocode(city)
    if location:
      lat_lon.append(location)
  except GeocoderTimedOut as e:
      print("Error: geocode failed on input %s with message %s">
            (city, e))
lat_lon
```

```python
df = pd.DataFrame(lat_lon, columns=['City Name', 'Coordinates'])
df.head(7)
```

```python
geometry = [Point(x[1], x[0]) for x in df['Coordinates']]
geometry[:7]
```

```python
#set coordinate system
crs = {'init': 'epsg:4326'}

#converts df to geo df
geo_df = gpd.GeoDataFrame(df, crs=crs, geometry=geometry)
geo_df.head()
```
